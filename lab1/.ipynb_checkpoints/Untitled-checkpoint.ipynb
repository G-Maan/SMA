```python
import os
import pandas as pd
import folium
from folium import plugins
from folium.plugins import HeatMap
```

```python
topHashtags = pd.read_excel("export_content_x_uk_x_filter_nativeretweets_geocode_55_378051_3_43597299999999_750km_2016_04_21_11_46_43.xlsx", sheet_name="Top Hashtags")
```

```python
tweets = pd.read_excel("export_dashboard_x_uk_x_filter_nativeretweets_geocode_55_378051_3_43597299999999_750km_2016_04_21_10_32_03.xlsx", "Stream", usecols="B,C,G,J,K,O,R").dropna()
```

```python
import numpy as np
from folium.plugins import HeatMap

m = folium.Map([51.509865, -0.118092], zoom_start=6)
hashtagArr = tweets[['Latitude', 'Longitude']].values
m.add_child(HeatMap(hashtagArr, radius=15))
m.save('Heatmap.html')
m
```

```python
def createMap(data, hashtag, i):
    hashtagData = data[data['Tweet content'].str.contains(hashtag)]
    
    m = folium.Map([51.509865, -0.118092], zoom_start=6)
    
    hashtagData = [[i, j] for i, j in zip(hashtagData['Latitude'], hashtagData['Longitude'])]
    print(hashtagData)
    HeatMap(hashtagData).add_to(m)
    
    m.save("maps/" + str(i+1) + hashtag + ".html")
```

```python
tags = topHashtags.iloc[:5]
for i, hashtag in enumerate(tags['Hash']):
    createMap(tweets, hashtag, i)
```
